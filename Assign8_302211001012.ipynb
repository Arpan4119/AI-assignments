# Problem Statement
Write rules in Prolog to determine the best move in Tic-Tac-Toe for any given board configuration. Assume that the position of pieces is given by a predicate 'p'.



```
% Define the Tic-Tac-Toe board size (3x3).
board_size(3).

% Predicate to check if a cell is empty.
empty_cell(Row, Col) :-
    board_size(Size),
    between(1, Size, Row),
    between(1, Size, Col),
    \+ p(_, Row, Col).

% Predicate to check if a player has won.
win(Player, Row, Col) :-
    board_size(Size),
    (   % Check rows
        forall(between(1, Size, C), p(Player, Row, C))
    ;   % Check columns
        forall(between(1, Size, R), p(Player, R, Col))
    ;   % Check diagonal from top-left to bottom-right
        (Row = Col, forall(between(1, Size, I), p(Player, I, I)))
    ;   % Check diagonal from top-right to bottom-left
        (Row + Col =:= Size + 1, forall(between(1, Size, I), p(Player, I, Size - I + 1)))
    ).

% Predicate to switch players.
other_player(x, o).
other_player(o, x).

% Predicate to find a winning move or a blocking move.
ttt_move(Player, Row, Col) :-
    empty_cell(Row, Col),
    (   % Check if the move results in a win.
        assert(p(Player, Row, Col)),
        win(Player, Row, Col),
        writeln('win!')
    ;   % Check if the opponent has a winning move, and block it.
        other_player(Player, Opponent),
        assert(p(Opponent, Row, Col)),
        win(Opponent, Row, Col),
        writeln('move to block opponent!')
    ;   % If neither win nor block, retract the move.
        retract(p(Player, Row, Col)),
        fail
    ).
```
