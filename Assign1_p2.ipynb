**Python program to load and read a numerical dataset from a CSV file and draw the corresponding graph considering the dataset as an adjacency matrix.**

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def load_weighted_digraph_from_adjacency_matrix(file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(file_path, index_col=0)

    # Convert the DataFrame to a NumPy array
    adjacency_matrix = df.to_numpy()

    # Get the number of nodes in the graph
    num_nodes = len(adjacency_matrix)

    # Create a weighted directed graph (digraph) 'g' using NetworkX
    g = nx.DiGraph()

    # Add nodes to the graph
    g.add_nodes_from(range(num_nodes))

    # Add edges based on the adjacency matrix with weights
    for i in range(num_nodes):
        for j in range(num_nodes):
            weight = adjacency_matrix[i, j]
            if weight > 0:
                g.add_edge(i, j, weight=weight)

    return g

def draw_weighted_digraph(g):
    # Extract edge weights as a dictionary
    edge_labels = nx.get_edge_attributes(g, 'weight')

    # Draw the graph with labels and edge weights
    pos = nx.circular_layout(g)
    nx.draw(g, pos, with_labels=True, node_size=1000, node_color="yellow", font_size=10, font_weight='bold', arrowsize=15, connectionstyle='arc3,rad=0.07')
    nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels, label_pos=0.25)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    # Replace 'graphCSV.csv' with the path to your CSV file
    csv_file_path = "/content/graph.csv"

    # Load the weighted digraph from the adjacency matrix in the CSV file
    weighted_digraph = load_weighted_digraph_from_adjacency_matrix(csv_file_path)

    # Draw the weighted digraph
    draw_weighted_digraph(weighted_digraph)
```

**Documentation**
