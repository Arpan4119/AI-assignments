# Code



```
% defining animals
animal(human).
animal(bat).
animal(parrot).
animal(duck).
animal(butterfly).
animal(spider).
animal(frog).
animal(salamander).
animal(crocodile).
animal(lizard).
animal(tuna).
animal(cod).

%defining mammals
is_mammal(human).
is_mammal(bat).

% defining birds
is_bird(parrot).
is_bird(duck).

% defining invertebrates
is_invertebrates(butterfly).
is_invertebrates(spider).

% defining amphibian
is_amphibian(frog).
is_amphibian(salamander).

% defing reptile
is_reptile(crocodile).
is_reptile(lizard).

% defining fish
is_fish(tuna).
is_fish(cod).

% defining animals which can swim
can_swim(duck).
can_swim(frog).
can_swim(crocodile).
can_swim(tuna).
can_swim(cod).
can_swim(human).

% defining animals which can fly
can_fly(bat).
can_fly(parrot).
can_fly(butterFly).

% defining animals which lives in water
lives_in_water(tuna).
lives_in_water(cod).

% Predicate to identify animals based on characteristics
find_animal(Characteristics, Animal) :-
    animal(Animal),
    satisfies_characteristics(Animal, Characteristics).

satisfies_characteristics(Animal, Characteristics) :-
    \+ (member(Characteristic, Characteristics), \+ has_characteristic(Animal, Characteristic)).

has_characteristic(Animal, Characteristic) :-
    (is_mammal(Animal), Characteristic = 'mammal');
    (is_bird(Animal), Characteristic = 'bird');
    (is_invertebrates(Animal), Characteristic = 'invertebrates');
    (is_fish(Animal), Characteristic = 'fish');
    (is_reptile(Animal), Characteristic = 'reptile');
    (is_amphibian(Animal), Characteristic = 'amphibian');
    (can_fly(Animal), Characteristic = 'fly');
    (can_swim(Animal), Characteristic = 'swim');
    (lives_in_water(Animal), Characteristic = 'aquatic').

```



This code is a Prolog program that defines a database of animals and their characteristics, and it provides a predicate called `find_animal` that allows you to query for animals based on certain characteristics. Here's a short note on the key components of the code:

5. **Usage**: You can use the `find_animal` predicate to find animals with specific characteristics. For example, if you query `find_animal(['mammal', 'swim'], Animal).`, it would return `Animal = human`, indicating that humans are mammals that can swim.
